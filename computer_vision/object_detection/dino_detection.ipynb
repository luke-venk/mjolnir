```python
from groundingdino.util.inference import load_model, load_image, predict, annotate
import cv2, torch
device = torch.device("cpu")

model = load_model("groundingdino/config/GroundingDINO_SwinT_OGC.py", "weights/groundingdino_swint_ogc.pth",  device=device
)
```

```python

image_name = '2844'
IMAGE_PATH = f".asset/{image_name}.jpg"
TEXT_PROMPT = "silver shiny sphere."
BOX_TRESHOLD = 0.35
TEXT_TRESHOLD = 0.25

image_source, image = load_image(IMAGE_PATH)

boxes, logits, phrases = predict(
    model=model,
    image=image,
    caption=TEXT_PROMPT,
    box_threshold=BOX_TRESHOLD,
    text_threshold=TEXT_TRESHOLD, device = device
)


# annotated_frame = annotate(image_source=image_source, boxes=boxes, logits=logits, phrases=phrases)
# cv2.imwrite("annotated_image.jpg", annotated_frame)
```

```python
import numpy as np
idx = np.argmin(np.abs((boxes[:,2]-boxes[:,0])/(boxes[:,3]-boxes[:,1]) - 1))
boxes  = boxes[idx:idx+1]
logits = logits[idx:idx+1]
phrases = [phrases[idx]]
annotated_frame = annotate(
    image_source=image_source,
    boxes=boxes,
    logits=logits,
    phrases=phrases
)

cv2.imwrite(f"annotated_{image_name}.jpg", annotated_frame)
```
